```python
# Range of IDS
min_id = 200
max_id = 205  
```

```python
player_response = {}
```

```python
for player_id in range(min_id, max_id):
    querystring = {"id": str(player_id), "season": "2022"}
    response = requests.get(url, headers=headers, params=querystring)
    player_data = response.json()
    player_response[player_id] = player_data
```

```python
import pandas as pd

df = pd.DataFrame(player_table_data, columns=[
    'PlayerID',
    'PlayerLastName',
    'PlayerFirstName',
    'Season',
    'Posistion',
    'Nationality',
    'Age',
    'Height',
    'Weight',
    'ShotsOnGoal',
    'TotalShots',
    'TotalGoals',
    'Assists',
    'Conceded',
    'Saves',
    'TotalPasses',
    'Accuracy',
    'Tackles',
    'Blocks',
    'Interceptions',
    'DribblesAttempts',
    'DribbleSuccess',
    'DribblePast',
    'DuelsWon',
    'DuelsTotals',
    'FoulsDrawn',
    'FoulsCommitted',
    'YellowCards',
    'RedCards',
    'GameExpulsions'])
```
