```python
# importing Dependencies
import requests
import json
import pandas as pd
```

```python
url = "https://api-football-v1.p.rapidapi.com/v3/fixtures"

querystring = {"league":"371","season":"2023"}

headers = {
	"x-rapidapi-key": "e745f20ff6mshb3a174087c1d159p11ee9ejsn4221f696691e",
	"x-rapidapi-host": "api-football-v1.p.rapidapi.com"
}

response_team = requests.get(url, headers=headers, params=querystring)

print(json.dumps(response_team.json(), indent=4))
response_team_j = response_team.json()
```

```python
# List to store fixture data
fixtures_data = []
    
# Iterate through the fixtures and extract required information
for fixture in response_team_j['response']:
    fixture_data = {
        "Fixture ID": fixture['fixture']['id'],
        "Referee": fixture['fixture'].get('referee', 'N/A'),
        "Dates": fixture['fixture']['date'],
        "Venue": fixture['fixture']['venue']['name'],
        "City": fixture['fixture']['venue']['city'],
        "Home Team": fixture['teams']['home']['name'],
        "Home Team Logo": fixture['teams']['home']['logo'],
        "Away Team": fixture['teams']['away']['name'],
        "Away Team Logo": fixture['teams']['away']['logo'],
        "Goals Home": fixture['goals']['home'],
        "Goals Away": fixture['goals']['away']
    }
        # Add fixture data to the list
    fixtures_data.append(fixture_data)
```

```python
# Create DataFrame from the list of fixtures
df_fixtures = pd.DataFrame(fixtures_data)
```

```python
#dispolay dataframe
df_fixtures.head()
```

```python
# Convert DateTime column to datetime
df_fixtures['Dates'] = pd.to_datetime(df_fixtures['Dates'])

# Extract Date and Time
df_fixtures['Date'] = df_fixtures['Dates'].dt.date
df_fixtures['Time'] = df_fixtures['Dates'].dt.time

# Drop the original DateTime column
df_cleaned = df_fixtures.drop(columns=['Dates'])
```
